```python
import tqdm
from time import sleep
```

#### 기본 활용 : tnrange == tqdm_notebook + range -> range(4) 와 동일하게 사용

```python
for i in tqdm.tnrange(4):
    for j in tqdm.tnrange(100):
        sleep(0.01)
```

#### trange == tqdm + range -> 콘솔에서는 progress bar 로 시각화

```python
for i in tqdm.trange(4):
    for j in tqdm.trange(100):
        sleep(0.01)
```

```python
for i in tqdm.tqdm(range(4)):
    for j in tqdm.tqdm(range(100), mininterval=3):
        sleep(0.1)
```

```python
for i in tqdm.tgrange(4):
    for j in tqdm.tgrange(100):
        sleep(0.01)
```

#### Pandas dataframe 에 apply 함수를 적용할 때 progress bar 를 보면서 체크가 가능하다

```python
import pandas as pd
import numpy as np
import tqdm

df = pd.DataFrame(np.random.randint(0, 100, (100000, 100)))
tqdm.tqdm.pandas(ncols=100)  # can use tqdm_gui, optional kwargs, etc`

# Now you can use `progress_apply` instead of `apply`
df.groupby(0).progress_apply(lambda x: x**2)
```
