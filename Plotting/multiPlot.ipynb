```python
# Libraries
import matplotlib.pyplot as plt
import pandas as pd
from math import pi
import numpy as np
 
# Set data
df = pd.DataFrame({
'group': ['A','B','C','D'],
'var1': [38, 1.5, 30, 4],
'var2': [29, 10, 9, 34],
'var3': [8, 39, 23, 24],
'var4': [7, 31, 33, 14],
'var5': [28, 15, 32, 14]
})
 
# number of variable
categories=list(df)[1:]
N = len(categories)
 
# We are going to plot the first line of the data frame.
# But we need to repeat the first value to close the circular graph:

values=[]
for i in range(len(df["group"])):
    temp=df.loc[i].drop('group').values.flatten().tolist()
    temp+=temp[:1]
    values.insert(i,temp)

 
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 

```

```python
def multi_plot(figure, f_num):
    figure.ax = plt.subplot(2, 2, f_num, polar=True)

    plt.xticks(angles[:-1], categories, color='grey', size=8)
    plt.yticks([10,20,30], ["10","20","30"], color="grey", size=7)
    plt.ylim(0,40)

    figure.ax.set_rlabel_position(0) 
    figure.ax.plot(angles, values[f_num-1], linewidth=1, linestyle='solid')
    figure.ax.fill(angles, values[f_num-1], 'b', alpha=0.1)
```

```python
f = plt.figure(figsize=(10,10))


multi_plot(f, 1)
multi_plot(f, 2)
multi_plot(f, 3)
multi_plot(f, 4)

f.show()
```
